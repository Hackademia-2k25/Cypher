```python
import os
from dotenv import load_dotenv
load_dotenv()

```

```python
from langchain_community.document_loaders import PyPDFLoader
pdf = 'material/rayoptics.pdf'

loader = PyPDFLoader(pdf)
doc = loader.load()
doc
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
# Initialize the splitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=3000,  # Max token size per chunk
    chunk_overlap=500,  # Overlap for context retention
    separators=["\n\n", "\n", ".", " "]  # Split by paragraphs, then sentences
)

# Apply to text
chunks = splitter.split_documents(doc)
chunks
```

```python
metadatas = [{"source": f"chunk_{i}"} for i in range(len(chunks))]
metadatas
```

```python
# Import necessary libraries
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.text_splitter import RecursiveCharacterTextSplitter

# 1. Initialize the embedding model through LangChain
embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2",
    model_kwargs={'device': 'cpu'},
    encode_kwargs={'normalize_embeddings': True}
)

# 2. Create the vector store using Chroma
# Assuming you have your chunks in text_chunks list
vectordb = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory="./chroma_db"  # This will store your embeddings locally
)

```

```python
# Load the existing database
loaded_vectordb = Chroma(
    persist_directory="./chroma_db",
    embedding_function=embeddings
)

```

```python
# Simple similarity search
query = "what is snells law?"
docs = loaded_vectordb.similarity_search(query, k=3)  # k is number of results

# Or with scores
docs_and_scores = loaded_vectordb.similarity_search_with_score(query, k=4)
```
